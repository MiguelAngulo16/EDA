```python
class nodo:
  def __init__(self,dato):
    self.dato = dato
    self.siguiente = None
```

```python
class lista:
  def __init__(self):
    self.cabeza = None
    self.tamano = 0

  def imprimir(self):
    nodo_recorre = self.cabeza
    while (nodo_recorre):
      print(nodo_recorre.dato)
      nodo_recorre = nodo_recorre.siguiente
    print("\n")

  def listaVacia(self):
    if self.cabeza == None:
      return True
    else:
      return False
    
  def contarElementos(self):
    print("La cantidad de elementos de la lista es de:",self.tamano)

  def insertarInicio(self, dato):
    nodo_nuevo = nodo(dato)
    nodo_nuevo.siguiente = self.cabeza
    self.cabeza = nodo_nuevo
    self.tamano = self.tamano + 1
    
  def insertarFinal(self, dato):
    nodo_nuevo = nodo(dato)
    recorre = self.cabeza
    if(self.listaVacia()):
      nodo_nuevo = nodo(dato)
      nodo_nuevo.siguiente = self.cabeza
      self.cabeza = nodo_nuevo
      self.tamano = self.tamano + 1
    else:
      while (recorre.siguiente != None):
        recorre = recorre.siguiente
      recorre.siguiente = nodo_nuevo
      self.tamano = self.tamano + 1

  def insertarIdx(self,dato,idx):
    if(idx == 0):
      self.insertarInicio(dato)
    elif(idx > 0 and idx <= (self.tamano - 1)):      
      cont = 0
      recorre = self.cabeza
      nodo_nuevo = nodo(dato)
      while(cont != (idx-1) and recorre.siguiente != None):
        recorre = recorre.siguiente
        cont += 1
      nodo_nuevo.siguiente = recorre.siguiente
      recorre.siguiente = nodo_nuevo
      self.tamano += 1
    else:
      print("El índice no existe en la lista.")

  def eliminarInicio(self):
    if(self.listaVacia()):
      print("La lista está vacia")
    else:
      self.cabeza = self.cabeza.siguiente
      self.tamano = self.tamano - 1

  def eliminarFinal(self):
    if(self.listaVacia()):
      print("La lista está vacia")
    else:
      recorre = self.cabeza
      while (recorre.siguiente.siguiente != None):
        recorre = recorre.siguiente
      recorre.siguiente = None
      self.tamano = self.tamano - 1

  def eliminarIdx(self,idx):
    if(idx == 0):
      self.eliminarInicio()
    elif(idx > 0 and idx <= (self.tamano - 1)):
      cont = 0
      recorre = self.cabeza
      while(cont != (idx-1) and recorre.siguiente != None):
        recorre = recorre.siguiente
      self.tamano -= 1
    else:
      print("El indice no existe en la lista.")

  def buscarDato(self,valor):
    recorre = self.cabeza
    while(recorre.dato != valor and recorre.siguiente != None):
      recorre = recorre.siguiente
    if(recorre.dato == valor):
      return True
    elif(recorre.dato != valor):
      return False
```

```python
l1 = lista()

for i in range(0,6):
    l1.insertarFinal(i)
l1.imprimir()

l1.insertarIdx(10,2)
l1.imprimir()

l1.eliminarIdx(2)
l1.imprimir()
```
